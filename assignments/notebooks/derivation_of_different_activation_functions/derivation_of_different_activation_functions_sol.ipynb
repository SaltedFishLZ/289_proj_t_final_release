# Derivatives of Different Activation Functions

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
def ReLU(z):
############TODO################
    return np.maximum(0,z)
############TODO################
def der_ReLU(z):
############TODO################
    return (z>0).astype('float')
############TODO################
def sigmoid(z):
############TODO################
    return 1 / (1 + np.exp(-z))
############TODO################
def der_sigmoid(z):
############TODO################
    return sigmoid(z) * (1-sigmoid(z))
############TODO################
def tanh(z):
############TODO################
    return np.tanh(z)
############TODO################
def der_tanh(z):
############TODO################
    return 1 - np.square(np.tanh(z))
############TODO################
def swish(z):
    beta=0.5
############TODO################
    return z / (1 + np.exp(-beta*z))
############TODO################
def der_swish(z):
    beta=0.5
############TODO################
    return (1 + np.exp(-beta*z) + beta * z * np.exp(-beta*z))/(1 + np.exp(-beta*z))**2
############TODO################
```

```python
def plot(f, df, yrng, xrng,name):
    z = np.arange(-100, 100, .1)
    y = f(z)
    y2 = df(z)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    b, = ax.plot(z, y, label='f')
    o, = ax.plot(z, y2, label='df')
    ax.set_ylim(yrng)
    ax.set_xlim(xrng)
    ax.grid(True)
    ax.set_xlabel('z')
    ax.set_title(name)
    plt.legend(handles=[b,o])
    plt.show()
plot (ReLU, der_ReLU, [-2.0, 2.0], [-2.0, 2.0],'ReLU')
plot (sigmoid, der_sigmoid, [-0.5, 1.5], [-10.0, 10.0],'sigmoid')
plot (tanh, der_tanh, [-2.0, 2.0], [-10.0, 10.0], 'tanh')
plot (swish, der_swish, [-2.0, 2.0], [-10.0, 10.0],'swish')
```

#### Start Discussion

You should see that for sigmoid-like activation functions (sigmoid and tanh), the derivatives are relatively small compared with ReLU. e.g. The maximum of derivative for sigmoid happens at z=0, and it is only 0.25, whereas ReLU's derivative is constant 1 when z>=0. This makes sigmoid-like activation functions prone to the gradient vanishing problem. Also, you should see that the first derivative of swish is a smoother version of the first derivative of ReLU.
